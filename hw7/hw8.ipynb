```python
from pubmed_lookup import PubMedLookup
from pubmed_lookup import Publication
from sklearn.feature_extraction.text import CountVectorizer

import pandas as pd
```

```python
email = ''
month = []
year = []
title = []
num = 0
num1 = 0
while True:
    if num1 == 500:
        break
    try:
        url = 'http://www.ncbi.nlm.nih.gov/pubmed/'+str(22551378+num)
        num+=1
        lookup = PubMedLookup(url, email)
        publication = Publication(lookup)    # Use 'resolve_doi=False' to keep DOI URL
        if publication.month is None or publication.year is None or publication.title is None:
            continue
        else:
            month.append(publication.month)
            year.append(publication.year)
            title.append(publication.title)
            num1+=1
            print(num1)
    except:
        print('cannot get document summary')
        num+=1
        
data = pd.DataFrame({'month':month, 'year': year,'title':title})    
data.to_csv('500_article', index=False)
```

```python
vectorizer = CountVectorizer(min_df = 1,stop_words = 'english')

document_term_matrix = vectorizer.fit_transform(data.title.to_list())

pd.DataFrame(document_term_matrix.toarray(),index =data.title.to_list(),columns = vectorizer.get_feature_names()).head(10)
```

```python
from rake_nltk import Metric,Rake

for i in data.title.to_list():
    r = Rake()
    r.extract_keywords_from_text(i)
    b = r.get_ranked_phrases()
    print(b)   #task2
```

```python
import pyfpgrowth
from nltk.corpus import stopwords

set(stopwords.words('english'))

stop_words = set(stopwords.words('english'))

transactions = []
for i in data.title.to_list():
    temp = []
    for word in i.split():
        if word not in stop_words:
            temp.append(word)
    transactions.append(temp)

patterns = pyfpgrowth.find_frequent_patterns(transactions,9)

print(patterns)
```

```python
from prefixspan import PrefixSpan

ps = PrefixSpan(transactions)
print(ps.frequent(9))
print('---------')
print(ps.frequent(8))
print('---------')
print(ps.frequent(7))
print('---------')
```
