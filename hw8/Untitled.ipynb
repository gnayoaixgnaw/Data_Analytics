```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from yellowbrick.text import TSNEVisualizer
from yellowbrick.text import UMAPVisualizer


data = pd.DataFrame()
data1 = pd.read_csv('Obesity.csv')
data1['label'] = data1['Title'].map(lambda x : 'Obesity' if x is not None else 'Obesity')
data1['text'] = data1['Title']+ data1['Abstract']

data2 = pd.read_csv('Cancer.csv')
data2['label'] = data2['Title'].map(lambda x : 'Cancer' if x is not None else 'Cancer')
data2['text'] = data2['Title']+ data2['Abstract']

data = data.append(data1,ignore_index = True)
data = data.append(data2,ignore_index = True)
data
```

```python
tfidf = TfidfVectorizer()

X = tfidf.fit_transform(data['text'].values.astype('U'))
Y = data['label']

# Create the visualizer and draw the vectors
tsne = TSNEVisualizer()
tsne.fit(X, Y)
tsne.show()
```

```python
umap = UMAPVisualizer()
umap.fit(X, Y)
umap.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA

ipca = IncrementalPCA(n_components=1)
ipca.fit(X)
IncrementalPCA(n_components=1)
ipca.transform(X) 
```
